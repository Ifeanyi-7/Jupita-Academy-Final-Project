BUILDING MACHINE LEARNING MODELS



```python
# copy dataframe into another dataframe
df2 = df.copy()
```

Skewness
check Skewness

```python
df2.skew()
```

```python
#Encoding
##Convert categorical variables in numerical variables
df2 = pd.get_dummies(df2)
df2
```

```python
# Divide dataset into x and y
x = df2.drop(['no_card'], axis=1)
y = df2[['no_card']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.2, random_state=42)
```

```python
 # using Support Vector Machine
from sklearn.svm import SVC

model = SVC()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
 #using Random Forest Classifier to check for a better accuracy score
from sklearn.ensemble import RandomForestClassifier

model2 = RandomForestClassifier()
model2.fit(x_train, y_train)
model2.score(x_test, y_test)
```

#Feature Scaling

```python
# Scale Features to improve accuracy scores of models
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()

scaled_features= [features for features in x.columns]
scaled_data= scaler.fit_transform(x)

x= pd.DataFrame(data=scaled_data, columns=scaled_features)
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.2, random_state=42)

model = SVC()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
model2 = RandomForestClassifier()
model2.fit(x_train, y_train)
model2.score(x_test, y_test)
```

```python
# Using model (Support Vector Machine)
y_pred = model.predict(x_test)
y_pred
```
